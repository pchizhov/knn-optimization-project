```python
import numpy as np
import matplotlib.pyplot as plt
from knn import KNNClassifier
from optimizer import LinearOptimizer
```

```python
positive = np.random.normal((2, 2), (2, 2), size=(200, 2))
negative = np.random.normal((-3, -3), (1, 1), size=(200, 2))
plt.scatter(np.concatenate((positive[:, 0], negative[:, 0])), np.concatenate((positive[:, 1], negative[:, 1])));
```

```python
classifier = KNNClassifier(10, LinearOptimizer)
classifier.fit(np.concatenate((positive, negative)), np.concatenate(([1] * 200, [0] * 200)))
classifier.predict(np.array([0, -1]))
```
